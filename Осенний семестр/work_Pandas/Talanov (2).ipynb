```python
import pandas as pd

#1

Xtr = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/X_train.csv'
studs = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/studs_info.csv'
Ytr = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/y_train.csv'
zach = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/zachety.csv'

x_train = pd.read_csv(Xtr)
studs_info = pd.read_csv(studs)
zachety = pd.read_csv(zach)
y_train = pd.read_csv(Ytr)

print(x_train.shape)
print(studs_info.shape)
print(zachety.shape)
print(y_train.shape)
```

```python
#2

train = pd.concat([x_train, y_train], axis=1)

train
```

```python
#5

train.columns = train.columns.str.lower()

train
```

```python
#6

train_h = pd.DataFrame(train.направление.str.split(' ', 1).tolist(), columns = ['код направления','название направления'])
train = pd.concat([train, train_h], axis=1)
train = train.drop('направление', axis=1)

train
```

```python
#7

studsh = studs_info[studs_info['Пол'] == "М"]
id = studsh['STD_ID'].values
trh = train.loc[~train['std_id'].isin(id)]
thrd = trh[trh['курс'] == 3.0]

thrd
```
