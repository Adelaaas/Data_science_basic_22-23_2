```python
#1.2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/home_work_clustering/segmentation%20data.csv'
table = pd.read_csv(url)

plt.figure(figsize=(10,6))
sns.displot(
    data=table.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
```

```python
#1.3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/home_work_clustering/segmentation%20data.csv'
table = pd.read_csv(url)

fig, axes = plt.subplots(1, 7, figsize=(15,6))

sns.boxplot(table['Sex'], ax=axes[0], width=0.3)
sns.boxplot(table['Marital status'], ax=axes[1], width=0.3)
sns.boxplot(table['Age'], ax=axes[2], width=0.3)
sns.boxplot(table['Education'], ax=axes[3], width=0.3)
sns.boxplot(table['Income'], ax=axes[4], width=0.3)
sns.boxplot(table['Occupation'], ax=axes[5], width=0.3)
sns.boxplot(table['Settlement size'], ax=axes[6], width=0.3)

```

```python
#1.5
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/home_work_clustering/segmentation%20data.csv'
table = pd.read_csv(url)

plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(table.corr(), xticklabels=table.corr().columns, yticklabels=table.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
