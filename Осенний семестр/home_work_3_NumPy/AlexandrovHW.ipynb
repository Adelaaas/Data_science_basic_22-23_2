```python
from IPython.core.interactiveshell import no_op
import numpy as np
n=int(input())
a=float(input())
d=float(input())
def asd(n,a,d):
  b=np.zeros((n,n))
  for i in range(n):
    b[i][i]=a+d*i
  return b
print(asd(n,a,d))
```
