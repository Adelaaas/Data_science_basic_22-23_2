```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]

# Первые 4 графика должны быть на одной figure, но на разных subplots
# 1 - обычный линей график y(x)

fig, ax = plt.subplots(2,2)
fig.set_figwidth(15)
fig.set_figheight(10)
ax[0,0].set_title('обычный линей график y(x)')
ax[0,0].plot(x,y,c="y")
ax[0,0].figsize=(15, 10)


# 2 - точечный график y(x)

ax[0,1].set_title('точечный график y(x)')
ax[0,1].scatter(x, y, c = 'g')
ax[0,1].figsize=(15, 10)

# 3 - гистограмма распределения x


ax[1,0].set_title('гистограмма распределения x')
ax[1,0].hist(x)
ax[1,0].figsize=(15, 10)

# 4 - гистограмма распределения y

ax[1,1].set_title('гистограмма распределения y')
ax[1,1].hist(y)
ax[1,1].figsize=(15, 10)
```

```python
# 5 и 6 графики должны быть на разных фигурах
# 5 - boxplot для x
# 6 - boxplot для y
fig, axes = plt.subplots(figsize=(15,10))  

sns.boxplot(x, ax=axes, width=0.3)


```

```python
fig, axes = plt.subplots(figsize=(15,6))  

sns.boxplot(y, ax=axes, width=0.3)
```

```python
# 7 график:
# создайте датафрейм со столбцами x, y, z
# постройте heatmap корреляции между x, y, z
data=pd.DataFrame({"x":x,"y":y, "z":z})

plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(data.corr(), xticklabels=data.corr().columns, yticklabels=data.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('heatmap корреляции между x, y, z', fontsize=22)
```
