```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]
```

 # Задания 1-4:

```python
#Задания 1,2,3,4.
plt.figure(figsize=(15, 10))   

fig, ax = plt.subplots(2,2)
fig.set_figwidth(15)
fig.set_figheight(10) 

ax[0,0].set_title("обычный линейный график y(x)")
ax[0,0].grid() 
ax[0,0].set_ylabel("y", fontsize=14)
ax[0,0].set_xlabel("x", fontsize=14)
ax[0,0].plot(x, y, c = 'r')


ax[0,1].set_title("точечный график y(x)")
ax[0,1].grid()  
ax[0,1].set_ylabel("y", fontsize=14)
ax[0,1].set_xlabel("x", fontsize=14)
ax[0,1].scatter(x, y, c = 'b') 

ax[1,0].set_title("гистограмма распределения x")
ax[1,0].grid() 
ax[1,0].hist(x)

ax[1,1].set_title("гистограмма расределения y")
ax[1,1].grid() 
ax[1,1].hist(y)
```

```python
df =pd.DataFrame({"x":x,"y":y, "z":z})
df.corr()
plt.figure(figsize=(14,12))
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('heatmap корреляции между x, y, z \n', fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
