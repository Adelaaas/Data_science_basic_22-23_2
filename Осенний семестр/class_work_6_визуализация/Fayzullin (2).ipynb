```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15,6))  
axes[0,0].plot(x,y)
axes[0,0].set_ylabel("y", fontsize=14)
axes[0,0].set_xlabel("x", fontsize=14)
axes[0,1].scatter(x,y)
axes[0,1].set_ylabel("y", fontsize=14)
axes[0,1].set_xlabel("x", fontsize=14)
axes[1,0].hist(x)
axes[1,0].set_ylabel("y", fontsize=14)
axes[1,0].set_xlabel("x", fontsize=14)
axes[1,1].hist(y)
axes[1,1].set_ylabel("y", fontsize=14)
axes[1,1].set_xlabel("x", fontsize=14)
```

```python
fig, ax = plt.subplots()
sns.boxplot(x, ax=ax, width=0.3)
plt.title('x box', fontsize=22)
```

```python
fig, ax = plt.subplots()
sns.boxplot(y, ax=ax, width=0.3)
plt.title('y box', fontsize=22)
```

```python
df = pd.DataFrame({"x" : x, "y": y, "z": z})
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)


plt.title('Heatmap', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
