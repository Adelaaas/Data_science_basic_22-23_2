```python
import pandas as pd
```

```python
print("Посмотреть размерность дата фреймов")

url1 = "https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/X_train.csv"
df1 = pd.read_csv(url1)
print("X_train:")
print(df1.shape)

url2 = "https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/studs_info.csv"
df2 = pd.read_csv(url2)
print("studs_info:")
print(df2.shape)

url3 = "https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/y_train.csv"
df3 = pd.read_csv(url3)
print("y_train:")
print(df3.shape)

url4 = "https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/zachety.csv"
df4 = pd.read_csv(url4)
print("zachety:")
print(df4.shape)
```

```python
print("Соедить два датафрейма X_train и y_train в один train по индексам")

pd.concat([df1, df3], axis=1)
```

```python
print("Посчитать пропуски в датафрейме studs_info")

df2.isna().sum()
```

```python
print("Посчитать сколько мужчин, сколько женщин в датафрейме studs_info")

df2['Пол'].value_counts()
```

```python
print("Создать новый датафрейм из studs_info, в котором будут только мужчины 2019-06-30 года выпуска")

df_res1 = pd.DataFrame(df2[df2['Пол'] == "М"])
df_res = pd.DataFrame(df_res1[df_res1['Дата выпуска'] == "2019-06-30"])
df_res
```

```python
print("Разложить колонку 'Дата выдачи' на четыре: Год выдачи, Месяц выдачи, День выдачи, Время выдачи")

df_res2 = pd.DataFrame(df2['Дата выдачи'].str.split('-',expand=True).values, columns=['Год выдачи', 'Месяц выдачи', 'Р'])
df_res3 = pd.DataFrame(df_res2['Р'].str.split(' ',expand=True).values, columns=['День выдачи', 'Время выдачи'])
df_res2.drop(['Р'], axis=1, inplace=True)

df_res_2 = pd.concat([df2, df_res2, df_res3], axis=1)
df_res_2
```

```python
print("Посчитать сколько уникальных значений в колонке studs_info['Статус'] и сколько раз каждое значение встречается в датафрейм")

print(df2["Статус"].nunique())
print(df2["Статус"].value_counts())
```

```python
print("Посчитать сколько раз какая оценка встречатся в датафрейм y_train")

df3['mark'].value_counts()
```

```python
print("Посчитать средний балл для студента с ID=175711 по всем предметам")

vsp = pd.concat([df1, df3], axis=1)

vsp2 = vsp[vsp['STD_ID'] == 175711]
vsp2['mark'].sum()/vsp2.shape[0]
```
