```python
import pandas as pd
```

```python
url1 = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/X_train.csv'
url2 = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/studs_info.csv'
url3 = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/zachety.csv'
url4 = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/work_Pandas/y_train.csv'

X_train = pd.read_csv(url1)
studs_info = pd.read_csv(url2)
zachety = pd.read_csv(url3)
y_train = pd.read_csv(url4)
```

```python
#Задание 1
print('Размерность X_train:', X_train.shape)
print('Размерность studs_info:', studs_info.shape)
print('Размерность zachety:', zachety.shape)
print('Размерность y_train:', y_train.shape)
```

```python
#Задание 2
train = pd.concat([X_train, y_train], axis=1)
train
```

```python
#Задача 3
print(studs_info.isna().sum().sum(), 'пустых значений в датафрейме')
```

```python
#Задача 4
print(len(train['АТТЕСТАЦИЯ'].unique()), 'уникальных значений в колонке АТТЕСТАЦИЯ')
print(train["АТТЕСТАЦИЯ"].value_counts())
```

```python
#Задача 5
train.rename(columns={'Unnamed: 0': 'unnamed: 0'}, inplace=True)
train.rename(columns={'STD_ID': 'std_id'}, inplace=True)
train.rename(columns={'НАПРАВЛЕНИЕ': 'направление'}, inplace=True)
train.rename(columns={'ГОД': 'год'}, inplace=True)
train.rename(columns={'КУРС': 'курс'}, inplace=True)
train.rename(columns={'СЕМЕСТР': 'семестр'}, inplace=True)
train.rename(columns={'АТТЕСТАЦИЯ': 'аттестация'}, inplace=True)
train.rename(columns={'ДИСЦИПЛИНА': 'дисциплина'}, inplace=True)
train
```

```python
#Задача 6
direct = train['направление'].str.split(expand = True)
direct[2] = direct[2].fillna('')
#Там сплит делил в том числе и названия направлений, поэтому надо фиксить циклом
train['Код направления'] = direct[0]
train['Название направления'] = direct[1] + ' ' + direct[2] 
train
```
