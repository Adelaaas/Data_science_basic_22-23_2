**Задача 1.**

**1-я входная строка** - год (опционально месяц и день) в формате ISO начала отсчёта.

**2-я входная строка** - год (опционально месяц и день) в формате ISO окончания отсчёта.

Создайте функцию, которая принимает на вход два введенных года и возвращает список (numpy array) дат с шагом в 1 день от начала до окончания отсчёта (последний день не включается).

Результат должен представлять из себя список дат в формате ISO.
Пример - 2005-02-25, где

- 2005 - год
- 02 - месяц
- 25 - день

````python
# ваше решение
import numpy as np


def beauty_output(num):
  return num if num >= 10 else '0' + str(num)
  

def dates(year1, year2):
  res = []
  month = 0
  for year in range(year1, year2):
    for month in range(1, 12 + 1):
      for date in range(1, 31 + 1):
        res.append(f"{year}-{ beauty_output(month) }-{ beauty_output(date) }")
  return res


year1 = int(input())
year2 = int(input())

dates(year1, year2)
````

**Задача 4.**

Вам дан массив, содержащий категории некоторого объекта в виде строк.

Создайте функцию, которая принимает на вход массив, хранящий категории объектов и возвращает:
- закодированный список из чисел, где каждому числу соответсвует какая-то категория;
- число категорий;
- словарь кодирования;

**Например:**

Ввод:

`('red', 'green', 'red', 'blue')`

Вывод:
```(0, 1, 0, 2)
Количество категорий: 3
Словарь кодирования: 
{'red': 0, 'green': 1, 'blue': 2'}
```

*P.s. встроенными функциями кодирования пользоваться запрещено.*

````python
# ваше решение


def my_split(line):
  res = []
  word = ''
  read_word = False
  for i in line:
    if i == "'":
      read_word = not read_word
      continue
    if read_word:
      word += i
    if not read_word and word != '':
      res.append(word)
      word = ''
  r1, r2, r3 = range(1, len(res) + 1, 1), 0, dict()
  for i in r1:
    r3[res[i - 1]] = 0
  for index, item in enumerate(r3.items()):
    r3[item[0]] = index 
  return r1, len(r3), r3


line = input()
r1, r2, r3 = my_split(line)
print(f"Количество категорий: {r2} \nСловарь кодирования: \n{r3}")
````
