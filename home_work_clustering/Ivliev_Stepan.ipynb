#2.1 Реализовать метод кластарезации KMeans: Методом локтя и силуета (в том числе с использованием графиков) оценить оптимальное число кластеров и обосновать его. Построить точечный график объектов до кластеризации и после с цветовым разделением на кластеры. Для построения графиков придется взять только две переменные как признаки, но строить кластеризацию не обязательно на двух переменных.

```python
from sklearn.cluster import KMeans

#ID,Sex,Marital status,	Age,	Education	Income,	Occupation,	Settlement size.
df = df.loc[:, ["ID", "Sex", "Marital status", "Age", "Education", "Income", "Occupation", "Settlement size"]]

SSE = []
plt.figure(figsize=(15, 8))
plt.title("Метод локтя", fontsize = 22)
for k in range(1, 10):
  kmeans = KMeans(n_clusters = k);
  kmeans.fit(df)
  SSE.append(kmeans.inertia_)

plt.plot(range(1,10), SSE, marker="s")
plt.xlabel("Clusters", fontsize = 20)
plt.ylabel("SSE", fontsize = 20)
plt.grid()
plt.show()
```

По графику видно, что за оптимальное количество кластеров k можно взять 4.

#Задание 2

```python
X = df.loc[:,['Age','Income']]
from sklearn.cluster import KMeans
n=4
kmeans = KMeans(n_clusters=n)
kmeans.fit(X)
Y_pred = kmeans.labels_
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df)
df['cluster'] = kmeans.labels_

fig, axes = plt.subplots(1, 2, figsize=(17,9))

fig.suptitle('Общий Заголовок',fontsize = 30)
axes[0].plot(df['Age'], df['Income'], 'ro')
axes[0].set_xlabel('Age',fontsize=19)
axes[0].set_ylabel('Income',fontsize=19)
axes[0].set_title("До кластеризации",fontsize=20)

axes[1].set_xlabel('Age',fontsize=19)
axes[1].set_ylabel('Income',fontsize=19)
axes[1].set_title("После кластеризации",fontsize=20)
axes[1].plot(df[df['cluster']==0]['Age'], df[df['cluster']==0]['Income'], 'bo', label='class1')
axes[1].plot(df[df['cluster']==1]['Age'], df[df['cluster']==1]['Income'], 'yo', label='class2')
axes[1].plot(df[df['cluster']==2]['Age'], df[df['cluster']==2]['Income'], 'ro', label='class3')
axes[1].plot(df[df['cluster']==3]['Age'], df[df['cluster']==3]['Income'], 'go', label='class4')

plt.legend(loc=0)
plt.show()
```

#Задание 3.2

```python
from sklearn.cluster import AgglomerativeClustering

Df = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
y_Df = Df.fit_predict(X)
X['pred_sklearn'] = y_Df

fig, ax = plt.subplots(1, 2, figsize=(16, 8), dpi = 80)

ax[0].set_title("До кластеризации",fontsize=20)
ax[0].set_xlabel("Age",fontsize=20)
ax[0].set_ylabel("Income",fontsize=20)
ax[0].scatter(df['Age'], df['Income'], s = 20)

ax[1].set_title("После кластеризации",fontsize=20)
ax[1].set_xlabel("Age",fontsize=20)
ax[1].set_ylabel("Income",fontsize=20)
ax[1].scatter(X.loc[y_Df==0, 'Age'], X.loc[y_Df==0, 'Income'], s=20, marker='o', color='orange')
ax[1].scatter(X.loc[y_Df==1, 'Age'], X.loc[y_Df==1, 'Income'], s=20, marker='o', color='black')
ax[1].scatter(X.loc[y_Df==2, 'Age'], X.loc[y_Df==2, 'Income'], s=20, marker='o', color='purple')
ax[1].scatter(X.loc[y_Df==3, 'Age'], X.loc[y_Df==3, 'Income'], s=20, marker='o', color='red')

plt.show()
```
