```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url='https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_2/main/home_work_clustering/segmentation%20data.csv'
df=pd.read_csv(url)
df
```

#1.1 Постройте Диаграмму разброса (точечный график) от двух переменных: Age и Income.

```python
Age =df['Age']
Income =df['Income']
Sex=df['Sex']
Marital_status=df['Marital status']
ID=df['ID']
Occupation=df['Occupation']
Settlement_size=df['Settlement size']
```

```python
plt.figure(figsize=(13, 13), dpi=50) #создаем фигуру
plt.title('Точечный график от переменных AGE и INCOME',fontsize=22) 
plt.grid() #сетка
plt.scatter(Age,Income, s=100,c='blue') #график от переменных
plt.xlabel('AGE',fontsize=25)
plt.ylabel('INCOME',fontsize=25)
```

 По графику нельзя определить зависимость между зарплатой клиента и его возрастом.


#1.2 Построить график анализа пропущенных значений.

```python
import seaborn as sns
plt.figure(figsize=(9,7))
sns.displot(
    data=df.isna().melt(value_name="ПРОПУСКИ"),
    y="variable",
    hue="ПРОПУСКИ",
    multiple="fill",
    aspect=3
)
plt.title('График анализа пропущенных значений',fontsize=20)
plt.ylabel('Variable',fontsize=15)
plt.xlabel('Count',fontsize=15)
```

По графику видно, что пропущенных значений нет.

#1.3 Построить boxplot для всех числовых переменных. График должен быть один, используя разные subplot() для каждой переменной (т.е. одна фигура и несколько сабплотов).

```python
plt.figure(figsize=(100,100))
fig, axes = plt.subplots(1, 2, figsize=(15,8))  

#числовые переменные - Age и Income (остальные либо 0 или 1, либо 0 или 1 или 2)
sns.boxplot(df['Age'], ax=axes[0], width=0.3)
axes[0].set_title("boxplot для Age",fontsize=18)
axes[0].grid()
axes[0].set_xlabel("Age", fontsize=14)

sns.boxplot(df['Income'], ax=axes[1], width=0.3)
axes[1].set_title("boxplot для Income",fontsize=18)
axes[1].grid()
axes[1].set_xlabel("Income", fontsize=14)
```

На boxplot можно выделить клиентов с зарплатой больше 200 000 и меньше 30 000,а также можно выделить клиентов старше 63(приблизительно).

#1.4 Построить barplot по переменной Sex.

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 7))
#df.Sex.value_counts()
sns.barplot(data = df, x = ['Женщины', 'Мужчины'] , y = [df.Sex.value_counts()[0], df.Sex.value_counts()[1]])
ax.set_title("Распределение полов")
plt.show()
```

Женщин больше чем мужчин

#Построить матрицу корреляции для всех переменных.

```python
df.corr()
plt.figure(figsize=(12,10), dpi= 100)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Матрица корреляции', fontsize=25)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.show()
```
