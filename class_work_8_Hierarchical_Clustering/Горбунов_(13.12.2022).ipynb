```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab import files
import pandas as pd


x = [11.5, 6, 11, 17, 17, 15.3, 15, 12, 5.4, 7.5]
y = [16.5, 15.5, 14.5, 10, 6.8, 5.6, 8, 7, 6, 4.3]

SSE = [] # Сумма квадратов ошибок для каждого результата

data = {'X': x, 'Y': y}
a = pd.DataFrame(data)



for k in range(1,9):
    kmeans = KMeans(n_clusters = k) # построить кластер
    kmeans.fit(a)
    SSE.append(kmeans.inertia_)
plt.plot(range(1,9), SSE, marker='s');
plt.title('Сlusters')
plt.xlabel('Number')
plt.ylabel('Error')
plt.show()
# наиболее подходящее кол-во кластеров 4
klasters=4
kmeans = KMeans(n_clusters=klasters)
kmeans.fit(a)
Y_pred = kmeans.labels_
a['cluster'] = Y_pred

plt.scatter(a[a['cluster']==0]['X'], a[a['cluster']==0]['Y'],c='red')
plt.scatter(a[a['cluster']==1]['X'], a[a['cluster']==1]['Y'],c='green')
plt.scatter(a[a['cluster']==2]['X'], a[a['cluster']==2]['Y'],c='blue')
plt.scatter(a[a['cluster']==3]['X'], a[a['cluster']==3]['Y'],c='yellow')

plt.title('Сoordinates')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()





```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab import files
#uploaded = files.upload()
import pandas as pd


x = [11.5, 6, 11, 17, 17, 15.3, 15, 12, 5.4, 7.5]
y = [6.5, 10, 14.5, 10, 6.8, 6, 8, 7, 6, 14.3]

data = {'X': x, 'Y': y}
a = pd.DataFrame(data)

import scipy.cluster.hierarchy as sch
#import dendrogram, linkage
from matplotlib import pyplot as plt
dendrogram = sch.dendrogram(sch.linkage(a,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()
# наиболее подходящие 4 кластера
klasters=4
kmeans = KMeans(n_clusters=klasters)
kmeans.fit(a)
Y_pred = kmeans.labels_
a['cluster'] = Y_pred

plt.scatter(a[a['cluster']==0]['X'], a[a['cluster']==0]['Y'], c='red')
plt.scatter(a[a['cluster']==1]['X'], a[a['cluster']==1]['Y'], c='yellow')
plt.scatter(a[a['cluster']==2]['X'], a[a['cluster']==2]['Y'], c='blue')
plt.scatter(a[a['cluster']==3]['X'], a[a['cluster']==3]['Y'], c='purple')
plt.title('Clasters')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()





```
