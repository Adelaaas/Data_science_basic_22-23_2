```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros((350, 350, 3))
```

```python
img.shape
```

```python
#0,0,255 = blue color
```

```python
img[:,:,:] = [0,0,255]
```

```python
img[0:70:,:,:] = [100,0,0]
img[280:350:,:,:] = [100,0,0]
img[70:280:,0:70,:] = [100,0,0]
img[70:280:,280:350,:] = [100,0,0]
```

```python
plt.imshow(img)
plt.show()
```

```python
chess = np.zeros((8, 8, 3))
chess[:,:,:] = [0,0,0]#black
chess[0::2, 0::2, :] = [255,255,255]#white
chess[1::2, 1::2, :] = [255,255,255]#white
```

```python
plt.imshow(chess)
plt.show()
```
