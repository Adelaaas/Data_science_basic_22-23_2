```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt
```

```python
array = np.arange(0, 129600*3, 1, np.uint8)
array = [0]*129600*3
array = np.reshape(array, (360, 360, 3))
```

```python
array[:, :, :] = [255, 0, 0]
array[:, :5, :] = [0, 0, 255]
array[:5, :, :] = [0, 0, 255]
array[:, 356:, :] = [0, 0, 255]
array[356:, :, :] = [0, 0, 255]
```

```python
plt.imshow(array, interpolation='none')
```
