```python

import cv2  # OpenCV
import matplotlib.pyplot as plt # Для визуализации
import numpy as np
from random import randint
     
```

```python
a = [[[32*i for i in range(3)] for i in range(700)] for i in range(700)]


img = np.asarray(a,dtype='uint8')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
a = [[[32*i for i in range(3)] for i in range(700)] for i in range(700)]

for x in range(700):
  for y in range(700):
    if x>=y:
      for i in range(3):
        a[x][y][i]*=2


img = np.asarray(a,dtype='uint8')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
a = [[[32*i for i in range(3)] for i in range(700)] for i in range(700)]

for x in range(700):
  for y in range(700):
    if x>=y:
      for i in range(3):
        a[x][y][i]*=2

for x in range(700):
  for y in range(700):
    if x-y>=y:
      for i in range(3):
        a[x][y][i]*=4

for x in range(700):
  for y in range(700):
    if y-x>=x:
      for i in range(3):
        a[x][y][i]-=30

img = np.asarray(a,dtype='uint8')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
