```python
df.loc[df["lumpy"]=='1' , 'lumpy'] = 0
df.loc[df["lumpy"]!=0 , 'lumpy'] = 1

df.loc[df["dominant_land_cover"]=='1' , 'dominant_land_cover'] = 1
df.loc[df["dominant_land_cover"]=='2' , 'dominant_land_cover'] = 2
df.loc[df["dominant_land_cover"]=='3' , 'dominant_land_cover'] = 3
df.loc[df["dominant_land_cover"]=='4' , 'dominant_land_cover'] = 4
df.loc[df["dominant_land_cover"]=='5' , 'dominant_land_cover'] = 5
df.loc[df["dominant_land_cover"]=='6' , 'dominant_land_cover'] = 6
df.loc[df["dominant_land_cover"]=='7' , 'dominant_land_cover'] = 7
df.loc[df["dominant_land_cover"]=='8' , 'dominant_land_cover'] = 8
df.loc[df["dominant_land_cover"]=='9' , 'dominant_land_cover'] = 9
df.loc[df["dominant_land_cover"]=='10' , 'dominant_land_cover'] = 10
df.loc[df["dominant_land_cover"]=='11' , 'dominant_land_cover'] = 11
df.loc[df["dominant_land_cover"]=='0' , 'dominant_land_cover'] = 0
df = df.astype(int)
```

```python
df.head()
```

```python
for i in df.columns:
  print(df[i].unique())
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['lumpy'], axis=1)
y = df['lumpy']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model = LogisticRegression()
model.fit(X_train, y_train)
answers_pred = model.predict(X_test)
```

```python
model.coef_
```

```python
d = model.coef_
b = df.columns
print('y =', round(model.intercept_[0], 3), end = '')
for i in range(1):
  print('+(', round(d[0][i], 3), ')*',b[i],sep ='', end =''  )
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

tn, fp, fn, tp = confusion_matrix(y_test, answers_pred).ravel()
Accuracy = (tn+tp)/(tp+tn+fp+fn)
Recall = (tp)/(tp+fn)
Precision = (tp)/(tp+fp)

print('Accuracy =', Accuracy)
print('Recall =', Recall)
print('Precision =', Precision)
print('f1 =', 2*Recall*Precision/(Recall+Precision))
```

```python
print(accuracy_score(y_test, answers_pred))
print(recall_score(y_test, answers_pred))
print(precision_score(y_test, answers_pred))
print(f1_score(y_test, answers_pred))
```
