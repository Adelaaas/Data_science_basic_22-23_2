```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
'''
Билет 13
Первое надо проверить данные на пропуски чтобы избавиться от строк ибо они не могут участвовать в математическом уравнении
Далее котегориальные признаки можно для каждого выделить новую колонку и поставить логическое значение пренадлежности 1 или 0 (да или нет)
Каждое уникальное значение получит свою колонку, в моем примере удалены все строковые данные
Далее разделить данные на тестовые(для проверки модели) и тренровочные(для получения коэффицентов при обучении модели)
Проверять модель можно только на данных не бывалых на обучении
Обучить модель и проверить ее ответы сравнив с правильными
Найти разницу между правдой и отданными ответами от модели и получить коэффицент ее правильности
Судя по коэффиценту можно сказать о 60% точности, однако она может быть повышена(а может и понижена) за счет
не учитывания признаков которые слабо влияют на результат(коэффиценты которых близки к нулю)
Также низкая точность оправдывается слабым влиянием всех признаков
'''
df = pd.read_csv("https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/%D0%92%D0%B5%D1%81%D0%B5%D0%BD%D0%BD%D0%B8%D0%B9%20%D1%81%D0%B5%D0%BC%D0%B5%D1%81%D1%82%D1%80/%D0%9A%D0%BE%D0%BB%D0%BB%D0%BE%D0%BA%D0%B2%D0%B8%D1%83%D0%BC/bestsellers%20with%20categories.csv")
# categorial Year Rating

df["User Rating"].unique()
df.Year.unique()
df


```

```python
df = df.drop(columns = 'Name', axis = 1)
df = df.drop(columns = 'Author', axis = 1)
#df = pd.get_dummies(df, columns=['Year'])
#df[df["Genre"]=="Non Fiction"] = 0 
#df[df["Genre"]=="Fiction"] = 1
df = pd.get_dummies(df, columns = ['Year', "User Rating", "Genre"])

df["Genre"]= df["Genre_Fiction"] - df["Genre_Non Fiction"]
df = df.drop(columns = 'Genre_Fiction', axis = 1)
df = df.drop(columns = 'Genre_Non Fiction', axis = 1)
df
```

```python
X = df.drop(columns = 'Genre', axis = 1)
y = df.Genre
X_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

model = LogisticRegression()

model.fit(X_train, y_train)

answ_pred = model.predict(x_test)
```

```python
print(accuracy_score(y_test,answ_pred))
```
