```python

import pandas as pd
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
```

```python
import numpy as np
import pandas as pd

columns = ['target'] + list(boston.feature_names)

df = pd.DataFrame(data=np.c_[boston.target, boston.data],
                  columns=columns)

df
```

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import seaborn as sns
df.corr()
plt.figure(figsize=(12,10), dpi= 100)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Матрица корреляции', fontsize=25)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.show()
```

```python
df_new = df.drop(columns = ['target'])
columns_name=df_new.columns.tolist()
x = np.array(df_new['RM']).reshape(-1, 1)
x
```

```python
columns_name
```

```python

y = np.array(df['target'])
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

print(f'Train dataset size: {X_train.shape}, {y_train.shape}')
print(f'Test dataset size: {X_test.shape}, {y_test.shape}')
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
w0 = model.intercept_
w0
```

```python
w1 = model.coef_
w1
```

```python
w0 = model.intercept_
w1 = model.coef_
result=""
result+=str(w0)+""
for i in range(len(w1)):
  result += " + "+str(w1[i])+" " + columns_name[i]+" "
result
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(x, y)

plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red', linewidth=2);
plt.show()
     
```

```python
l = (y_test - y_pred)**2
MSE_vruch = sum(l)/len(l)
MSE_vruch
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 score:', r2_score(y_test, y_pred))
```
