```python
from sklearn.datasets import load_boston

boston = load_boston()

print(boston.DESCR)
```

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
columns = ['target'] + list(boston.feature_names)

df = pd.DataFrame(data=np.c_[boston.target, boston.data],
                  columns=columns)

df
```

```python
y=np.array(df["target"])
y
```

```python
x1=(df.drop(["target"], axis=1))
x=np.array(x1["AGE"]).reshape(-1,1)
print(x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

print(f'Train dataset size: {X_train.shape}, {y_train.shape}')
print(f'Test dataset size: {X_test.shape}, {y_test.shape}')
```

```python
model = LinearRegression()
model.fit(X_train, y_train)  
print('Веса всех признаков (w1): ', model.coef_)
print('Свободный коэффицент уравнения w0:', model.intercept_)
```

```python
print("y =" ,  model.coef_[0],"+",model.intercept_,"AGE")
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red', linewidth=2);
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(x, y)

plt.show()  
```

```python
a=((y_test-y_pred)**2)
print(sum(a)/len(a))
```

```python


print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 score:', r2_score(y_test, y_pred))
```

```python
R2 >0 значит модель  работает лучше чем простое усреднение
MSE далек от 0 значит модель работает плохо
```
