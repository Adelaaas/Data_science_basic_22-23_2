```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

from google.colab import files
uploaded = files.upload()
```

```python
train = pd.read_csv('train.csv', sep=',')
train = train.dropna(axis = 0, how = 'any')
train
```

```python
tmp = np.array(map(lambda x: int(x.split('.')[1]), train['datetime']))
train['month'] = tmp 
hour = np.array(map(lambda x: int(x.split(' ')[1][:2]), train['datetime']))
train['Hour'] = hour
train
```

```python
Kitsuragi = LinearRegression()
x_train = train.drop(columns = ['total', 'datetime'])
y_train = train['total']
Kitsuragi.fit(x_train, y_train)
```

```python
sample = pd.read_csv('sample.csv', sep = ',')
```

```python
vsp = np.array(map(lambda x: int(x.split('.')[1]), sample['datetime']))
sample['month'] = vsp 
hour = np.array(map(lambda x: int(x.split(' ')[1][:2]), sample['datetime']))
sample['Hour'] = hour
sample_x = sample.drop(columns = ['datetime', 'total'])
ans = Kitsuragi.predict(sample_x)
sample['total'] = ans
sample = sample.drop(columns = ['month', 'Hour'])
# Ответ на вопрос
sample
```
