```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv('train.csv', sep=',')
train = train.dropna(axis = 0, how = 'any')
train
```

```python
tmp = np.array(map(lambda x: int(x.split('.')[1]), train['datetime']))
train['month'] = tmp 
hour = np.array(map(lambda x: int(x.split(' ')[1][:2]), train['datetime']))
train['Hour'] = hour
train
```

```python
Kitsuragi = LinearRegression()
x_train = train.drop(columns = ['total', 'datetime'])
y_train = train['total']
Kitsuragi.fit(x_train, y_train)
```

```python
test = pd.read_csv('sample.csv', sep = ',')
```

```python
tmp = np.array(map(lambda x: int(x.split('.')[1]), test['datetime']))
test['month'] = tmp 
hour = np.array(map(lambda x: int(x.split(' ')[1][:2]), test['datetime']))
test['Hour'] = hour
test_x = test.drop(columns = ['datetime', 'total'])
ans = Kitsuragi.predict(test_x)
test['total'] = ans
test = test.drop(columns = ['month', 'Hour'])
test
```
