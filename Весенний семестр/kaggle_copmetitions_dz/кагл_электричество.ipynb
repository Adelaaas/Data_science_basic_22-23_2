```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
test = pd.read_csv('sample.csv')
train = pd.read_csv('train.csv')
```

```python
train = train.dropna()
```

```python
d = []
m = []
t = []
y = []
for i in train['datetime']:
  d.append(int(i.split(' ')[0].split('.')[0]))
  m.append(int(i.split(' ')[0].split('.')[1]))
  y.append(int(i.split(' ')[0].split('.')[2]))
  T = (i.split(' ')[1])
  t.append(int(T.split(':')[0]) * 3600 + int(T.split(':')[1])*60 + int(T.split(':')[2]))
print(t)
```

```python
train['year'] = y
train['month'] = m
train['time'] = t
train['day'] = d
train = train.drop(columns = ['datetime'])
```

```python
X  = train.drop(columns = ['total'])
y = train['total']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
Gendu = DecisionTreeRegressor(max_depth=10, random_state=1)
Gendu.fit(X_train, y_train)
y_pred_tree = Gendu.predict(X_test)
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred_tree))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred_tree))
print('R2 score:', r2_score(y_test, y_pred_tree))
```
