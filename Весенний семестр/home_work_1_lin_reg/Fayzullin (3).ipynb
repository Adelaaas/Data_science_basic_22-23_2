```python
#Ð£Ð´Ð°Ð»ÐµÐ½Ð¸Ðµ Ð²ÑÐµÑ… Ð½ÐµÑ‡Ð¸ÑÐ»Ð¾Ð²Ñ‹Ñ… Ñ…Ð°Ñ€Ð°ÐºÑ‚ÐµÑ€Ð¸ÑÑ‚Ð¸Ðº
X = Data_f.drop(['Attrition_rate', 'Relationship_Status', 'Employee_ID', 'Hometown', 'Unit', 'Gender', 'Decision_skill_possess', 'Compensation_and_Benefits'  ], axis=1)
y = Data_f['Attrition_rate']

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```
