```python
#Удаление всех нечисловых характеристик
X = Data_f.drop(['Attrition_rate', 'Relationship_Status', 'Employee_ID', 'Hometown', 'Unit', 'Gender', 'Decision_skill_possess', 'Compensation_and_Benefits'  ], axis=1)
y = Data_f['Attrition_rate']

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
w0 = model.intercept_
w1 = model.coef_
print('w0 =', w0)
print('Age =', w1[0])
print('Education_Level =' ,w1[1])
print('Time_of_service =' ,w1[2])
print('Time_since_promotion =' ,w1[3])
print('growth_rate =' ,w1[4])
print('Travel_Rate =' ,w1[5])
print('Post_Level =' ,w1[6])
print('Pay_Scale =' ,w1[7])
print('Work_Life_balance =' ,w1[8])
print('VAR1 =' ,w1[9])
print('VAR2 =' ,w1[10])
print('VAR3 =' ,w1[11])
print('VAR4 =' ,w1[12])
print('VAR5 =' ,w1[13])
print('VAR6 =' ,w1[14])
print('VAR7 =' ,w1[15])
w1.dtype
```

```python
#Можно заметить, что влияние на результат всех характеристик
print('y=', round(w0, 5),'+',round(w1[0], 5), '*Age','+',round(w1[1], 5), '*Education_Level','+',round(w1[2], 5), '*Time_of_service','+',round(w1[3], 5), '*Time_since_promotion','+',round(w1[4],5), '*growth_rate','+',round(w1[5], 5), '*Travel_Rate','+',round(w1[6], 5), '*Post_Level','+',round(w1[7], 5), '*Pay_Scale','+',round(w1[8], 5), '*Work_Life_balance','+',round(w1[9], 5), '*VAR1','+',round(w1[10], 5), '*VAR2','+',round(w1[11], 5), '*VAR3','+',round(w1[12], 5), '*VAR4','+',round(w1[13], 5), '*VAR5','+',round(w1[14], 5), '*VAR6', '+',round(w1[15], 5), '*VAR7', sep=' ')
```

```python
y_pred = model.predict(X_test)
import matplotlib.pyplot as plt

plt.scatter(X_test['growth_rate'], y_test)
plt.plot( X_test['growth_rate'] , y_pred  , color='red', linewidth=2);
plt.show()
```

```python
y_pred = model.predict(X_test)

print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 score:', r2_score(y_test, y_pred))
#По метрикам видно, что модель работает очень плохо, так как параметр R2 отрицателен, а MSE меньше единицы
```
