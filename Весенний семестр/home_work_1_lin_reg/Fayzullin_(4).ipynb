```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')

```

```python
Data_f=pd.read_csv('/content/drive/MyDrive/Train.csv')
Data_f.loc[Data_f['Gender'] == 'F', 'Gender'] = 0
Data_f.loc[Data_f['Gender'] == 'M','Gender'] = 1
Data_f.loc[Data_f['Relationship_Status'] == 'Single', 'Relationship_Status'] = 0
Data_f.loc[Data_f['Relationship_Status'] == 'Married','Relationship_Status'] = 1
Data_f = Data_f.dropna()
```

```python
#Удаление всех нечисловых характеристик
X = Data_f.drop(['Attrition_rate',  'Employee_ID', 'Hometown', 'Unit',  'Decision_skill_possess', 'Compensation_and_Benefits'  ], axis=1)
y = Data_f['Attrition_rate']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
w0 = model.intercept_
w1 = model.coef_
print('w0 =', w0)
print('Post_Level =' ,w1[0])
print('Work_Life_balance =' ,w1[1])
print('VAR3 =' ,w1[2])
print('VAR7 =' ,w1[3])
w1.dtype
a= []
```

```python
#Можно заметить, что влияние на результат всех характеристик очень  мало
print('y=', round(w0, 5), '+',round(w1[0], 5), '*Post_Level','+',round(w1[1], 5), '*Work_Life_balance','+',round(w1[2], 5), '*VAR3','+',round(w1[3], 5), '*VAR7', sep=' ')
```

```python
y_pred = model.predict(X_test)
plt.scatter(X_test['VAR3'], y_test)
plt.scatter( X_test['VAR3'] , y_pred  );
plt.ylabel("Attrition_rate", fontsize=14)
plt.xlabel("VAR3", fontsize=14)
plt.show()
```

```python


print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 score:', r2_score(y_test, y_pred))
#По метрикам видно, что модель работает очень плохо, так как параметр R2 отрицателе
```
