```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]
```

```python
plt.figure(figsize=(15, 10))   
fig, ax = plt.subplots(2,2)
fig.set_figwidth(15)
fig.set_figheight(10) 
ax[0,0].set_title("Задание 1")
ax[0,0].set_ylabel("y")
ax[0,0].set_xlabel("x")
ax[0,0].plot(x, y, c = 'y')
ax[0,1].set_title("Задание 2")
ax[0,1].set_ylabel("y")
ax[0,1].set_xlabel("x")
ax[0,1].scatter(x, y, c = 'g') 
ax[1,0].set_title("Задание 3")
ax[1,0].set_xlabel("x")
ax[1,0].hist(x)
ax[1,1].set_title("Задание 4")
ax[1,1].set_xlabel("y")
ax[1,1].hist(y)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].boxplot(x)
ax[0].set_ylabel("x")
ax[1].boxplot(y)
ax[1].set_ylabel("y")
plt.show()
```

```python
plt.figure(figsize=(15,10))
df =pd.DataFrame({"x":x,"y":y, "z":z})
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.show()
```
